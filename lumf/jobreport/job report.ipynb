```python
#参考https://blog.csdn.net/hohaizx/article/details/79101322
data = {"商城-ABTest" : [20,'green', 0],
              "商城-o2o" : [20,'green', 0],
              "商城-首页原生": [20,'green',0],
              "商城-搜索" : [20,'green', 0],
               "致良知": [20,'red', 0],
               "药诊店" : [20,'blue', 0],
             }    # 各部分标签

df_data = pd.DataFrame(data)
```

```python

def draw_pie(df): 
    """
    绘制饼图
    explode：设置各部分突出
    label:设置各部分标签
    labeldistance:设置标签文本距圆心位置，1.1表示1.1倍半径
    autopct：设置圆里面文本
    shadow：设置是否有阴影
    startangle：起始角度，默认从0开始逆时针转
    pctdistance：设置圆内文本距圆心距离
    返回值
    l_text：圆内部文本，matplotlib.text.Text object
    p_text：圆外部文本
    """
    colors = df.iloc[1]
    weights = df.iloc[0]
    labels = df.columns
    explode = df.iloc[2]
    
#     fig, axes = plt.subplots(figsize=(6,6)) # 设置绘图区域大小
    fig, axes = plt.subplots(figsize=(10,5),ncols=2) # 设置绘图区域大小
    ax1, ax2 = axes.ravel()
    
    patches, l_text, p_text = ax1.pie(weights, explode=explode, 
                                      colors=colors, labels=labels, labeldistance=1.1,
                                      autopct="%1.1f%%", 
                                      shadow=False, startangle=90, pctdistance=0.6)
#     plt.figure(figsize=(15, 10))
    ax1.axis('equal')
    ax1.set_title('人力投入比重', loc='center')

# ax2 只显示图例（legend）
    ax2.axis('off')
    ax2.legend(patches, labels, loc='center left')
    
    plt.tight_layout()
    plt.show()

draw_pie(df_data)
```
